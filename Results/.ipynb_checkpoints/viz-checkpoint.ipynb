```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def map_plot(df):
    plt.plot(df['disc_tr_post'], df['acc_tr_post'], label='train')
    plt.plot(df['disc_te_post'], df['acc_te_post'], label='test')
    plt.xlabel('Discrimination')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.show()
```

# COMPAS

### Correct?

```python
compas = pd.read_csv('compas-p1d2.csv')
compas = compas.groupby('epsilon').mean()
map_plot(compas)
```

### Incorrect

```python
compas = pd.read_csv('compas-p2d1.csv')
compas = compas.groupby('epsilon').mean()
map_plot(compas)
```
